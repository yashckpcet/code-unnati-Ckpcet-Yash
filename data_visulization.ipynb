```python
#subplots

#plot 1
x = np.array([1,2,3,4])
y = np.array([3,4,5,6])

#plot 2
a = np.array([1,0,9,5])
b = np.array([5,6,7,8])

plt.subplot(1, 2, 1)#1 for row 2 for columns and 1 for possition for graph
plt.plot(x,y)
plt.subplot(1, 2, 1)
plt.plot(a,b)
```

```python
plt.subplot(2, 1, 1)#1 for row 2 for columns and 1 for possition for graph
plt.plot(x,y)
plt.subplot(2, 1, 3)
plt.plot(a,b)
```

```python
plt.subplot(5, 2, 1)#1 for row 2 for columns and 1 for possition for graph
plt.plot(x,y)
plt.subplot(5, 2, 2)
plt.plot(a,b)
plt.subplot(5, 2, 3)
plt.plot(a,x)
plt.subplot(5, 2, 4)
plt.plot(a,y)
plt.subplot(5, 2, 5)
plt.plot(b,x)
plt.subplot(5, 2, 6)
plt.plot(b,y)
plt.subplot(5, 2, 7)
plt.plot(x,a)
plt.subplot(5, 2, 8)
plt.plot(x,b)
plt.subplot(5, 2, 9)
plt.plot(y,a)
plt.subplot(5, 2, 10)
plt.plot(y,b)
```
